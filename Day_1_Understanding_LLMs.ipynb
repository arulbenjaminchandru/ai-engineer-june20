# Section 2 - What is an LLM?

**What is it?**
A Large Language Model is a neural network trained on a very large amount of text. Its single core skill is deceptively simple: **given some text, predict the next token (a chunk of text).** Repeat that prediction over and over and you get sentences, code, summaries, and answers.

**Why does it matter?**
Almost everything you'll build - chatbots, RAG assistants, agents, copilots - sits on top of this one prediction engine. If you understand next-token prediction, the rest of the course clicks into place.

**How does it work? (at a glance)**
1. Your text is split into **tokens**.
2. The model reads all those tokens (its **context**).
3. It produces a probability for *every possible next token*.
4. A token is chosen, added to the text, and the process repeats until it stops.

**When will I use this?**
Every single time you call an LLM API or type into a chat assistant.

### Three properties to memorise
- **Stateless by default** - the model does not remember previous conversations on its own. Anything it should "remember" must be sent again as text.
- **Context-driven** - its answer depends entirely on the text you give it right now.
- **Probabilistic** - it deals in likelihoods, not certainties. That's *why* the same prompt can give different answers.

> **Accuracy note (important nuance):** The line "LLMs don't *know* facts, they just predict tokens" is a useful simplification - but be precise. During training the model *does* absorb a great deal of factual and world knowledge, stored implicitly in its billions of parameters (its "weights"). The honest statement is: **an LLM has no live database it looks things up in; it reconstructs answers from learned statistical patterns, so it can be confidently wrong.** That nuance is what separates a beginner from an engineer.

```mermaid
flowchart LR
    A[Your text] --> B[Tokenizer]
    B --> C[LLM reads context]
    C --> D[Probabilities for next token]
    D --> E[Pick a token]
    E --> F[Append to text]
    F --> C
    F --> G[Stop -> Final response]
```

---

### Instructor Explanation
*"Open your phone's keyboard and start a sentence - 'I am going to the...' - and watch the three suggested words appear above the keys. Your phone is doing a tiny version of what an LLM does: predicting the next word from patterns. Now imagine that predictor trained on a large slice of the internet, with a memory of thousands of words instead of three. That's an LLM. Same trick, astonishing scale."*

### Remember This
> An LLM is a next-token prediction engine. It doesn't look facts up - it reconstructs them from patterns, which is why it can sound certain and still be wrong.

### Enterprise Perspective
A bank's customer-support assistant is *not* one giant brain. It's an LLM (the prediction engine) wrapped in guardrails: retrieval of approved policy documents, prompt rules ("answer only from the context"), and low-randomness settings. The LLM supplies fluent language; the **architecture around it** supplies the safety and the facts.

# Section 3 - Tokens (the unit the model actually reads)

**What is it?**
A **token** is a chunk of text the model processes - often a whole short word, sometimes part of a word, sometimes just punctuation or a space. Models don't read letters or words directly; they read tokens.

**Why does it matter?**
Three big reasons, and every AI engineer must know them cold:
- **Cost:** API pricing is *per token* (input + output). More tokens = more money.
- **Memory:** The **context window** is measured in tokens, not words. Tokens decide how much the model can "see" at once.
- **Performance:** Longer prompts cost more time and money and can dilute the model's focus.

**How does it work?**
A component called a **tokenizer** splits text into tokens using a learned vocabulary (commonly via an algorithm called Byte-Pair Encoding). Common words are usually one token; rare or long words get split into several.

**When will I use this?**
When you estimate cost, design prompts to fit a budget, or debug "context length exceeded" errors.

### Tokens are NOT words
A useful rule of thumb for English: **1 token ~ 4 characters ~ 0.75 words.** So **100 tokens ~ 75 words.** This varies by language and tokenizer, so always treat it as an estimate, not a law.

```
Text:    "Artificial Intelligence is amazing!"
Tokens:  ["Artificial", " Intelligence", " is", " amazing", "!"]
                 ^ note the leading spaces are part of the token
Count:   5 tokens  (for ~4 words)
```

> **Why a beginner gets surprised:** "tokenization" the *word* often becomes several tokens (e.g., `token` + `ization`), while a common word like "the" is a single token. So token count is not the same as word count - usually it's a bit higher.

---

### Instructor Explanation
*"Imagine you're paying for a taxi by the meter, not by the trip. It doesn't matter that you 'just want to go downtown' - the meter counts every unit of distance. Tokens are the LLM's meter. Every word you type, and every word it replies, ticks the meter. Once students feel that, they stop writing bloated prompts and start writing tight ones."*

### Remember This
> Tokens are the model's billing and memory unit. Words are for humans; tokens are for the meter. Roughly: 100 tokens ~ 75 English words.

### Enterprise Perspective
In a high-volume insurance assistant handling millions of queries a month, shaving 200 unnecessary tokens off a system prompt can cut a real five-figure sum from the monthly bill. Token discipline is cost engineering. Teams also cap output with **max tokens** so a runaway answer can't blow the budget.

### Interview Question
**Q: What is a token, and why does token count matter in production?**
> *Model answer:* "A token is the unit of text an LLM processes - often a word fragment, produced by a tokenizer. It matters because billing is per token, the context window is measured in tokens, and longer token counts increase latency and cost. Engineers manage prompts and outputs by token budget."

### Common Misconceptions
- **Wrong:** "1 word = 1 token."
  **Right:** "Tokens are often word fragments; ~100 tokens ~ 75 English words, and it varies by language."
- **Wrong:** "Only my output is billed."
  **Right:** "Both input (prompt + history) and output tokens are billed."

### Quick Check

2. **MCQ:** Roughly how many words are 1,000 tokens in English? (a) 250 (b) 750 (c) 1,500 (d) 4,000. *(b ~750.)*

````python
# Hands-On: a rough token estimator (NOT a real tokenizer, just intuition)
# Real tokenizers (e.g., tiktoken) are more accurate. Rule of thumb: ~4 chars/token.

def estimate_tokens(text):
    word_based = len(text.split())            # naive: words
    char_based = max(1, round(len(text) / 4)) # rule of thumb: ~4 chars per token
    return word_based, char_based

samples = [
    "AI is transforming software engineering",
    "Tokenization splits text into subword units",
    "The cat sat on the mat",
]

print(f"{'TEXT':45} | words | ~tokens")
print("-" * 70)
for s in samples:
    w, t = estimate_tokens(s)
    print(f"{s:45} | {w:5} | {t:7}")

print("\nNote: ~tokens is usually >= words because long words split into pieces.")
````

# Section 4 - Context Window (the model's short-term memory)

**What is it?**
The **context window** is the maximum number of **tokens** the model can consider at one time. Crucially, it includes *everything* in the request: your system prompt, the chat history, any retrieved documents, **and** the space reserved for the answer.

**Why does it matter?**
It explains one of the most common beginner questions: *"Why did the chatbot forget what I said earlier?"* When a conversation grows past the window, the oldest tokens fall out of view. The model isn't being forgetful - those tokens are simply no longer in the request.

**How does it work?**
```
+---------------------- CONTEXT WINDOW (token budget) -----------------------+
|  System prompt  |  Chat history (older -> newer)  |  Retrieved docs | Room |
|                 |  [oldest messages drop first]   |   (RAG)         | for  |
|                 |                                  |                 | reply|
+---------------------------------------------------------------------------+
   When total tokens exceed the window, content is truncated from the start.
```

**When will I use this?**
When designing long conversations, deciding how much history to keep, and choosing **RAG** instead of stuffing everything into the prompt.

### Why RAG helps
You cannot fit an entire company knowledge base into a context window, and doing so would be slow and expensive. **Retrieval-Augmented Generation (RAG)** fetches only the *few most relevant* chunks and places them in the window, so the model answers from fresh, specific, grounded text instead of from memory alone.

---

### Instructor Explanation
*"Picture a whiteboard of a fixed size. You can write the conversation on it, but when it fills up, you must erase the top to make room at the bottom. The model can only read what's currently on the whiteboard - not what you erased ten minutes ago. That's the context window. RAG is like a smart assistant who, instead of keeping everything on the board, runs to the filing cabinet and pins up only the one page you need right now."*

### Remember This
> The context window is the model's whiteboard, measured in tokens, shared by prompt + history + documents + answer. Run out of room and the oldest writing gets erased.

### Enterprise Perspective
An internal company knowledge assistant can't hold 10,000 policy pages in context. The production pattern is RAG: embed and index the documents, retrieve the top few relevant chunks per question, and feed only those into the window. This keeps answers grounded, cheap, and current - and lets you update knowledge by re-indexing, with no retraining.

### Interview Question
**Q: What is a context window, and how is it different from training data?**
> *Model answer:* "The context window is the live token budget the model can read for a single request - prompt, history, retrieved docs, and the reply. Training data is what shaped the model's weights once, ahead of time. Context is per-request and temporary; training is baked in and permanent. RAG works by putting the right text into the context window at query time."

### Common Misconceptions
- **Wrong:** "A bigger context window means the model remembers across separate sessions."
  **Right:** "It only affects a single request. Cross-session memory must be re-supplied as tokens."
- **Wrong:** "Context window is measured in words/messages."
  **Right:** "It's measured in tokens, and the reply shares the same budget."

### Quick Check
1. **True/False:** The context window includes the space for the model's answer. *(True.)*
2. **MCQ:** A chatbot 'forgets' earlier messages mainly because (a) a bug (b) older tokens dropped out of the window (c) the model is tired (d) the internet. *(b)*
3. **Scenario:** Users need answers from 5,000 internal PDFs. Stuff them all in the prompt or use RAG? *(RAG - retrieve only relevant chunks per question.)*


````python
# Hands-On: simulate context-window truncation
# A toy model with a tiny window so you can SEE older messages drop.

WINDOW = 20  # pretend the model can only hold 20 "tokens"

def tok(s):                      # toy tokenizer: 1 word = 1 token
    return s.split()

conversation = [
    ("system", "You are a helpful banking assistant."),
    ("user",   "Hi, what is my savings interest rate?"),
    ("assistant", "It is 4 percent per year."),
    ("user",   "And what about the fixed deposit rate please?"),
]

# Flatten to tokens (newest last), then keep only the last WINDOW tokens
all_tokens = []
for role, text in conversation:
    all_tokens += [f"[{role}]"] + tok(text)

kept = all_tokens[-WINDOW:]
dropped = all_tokens[:-WINDOW]

print(f"Total tokens: {len(all_tokens)} | Window: {WINDOW}")
print(f"\nDROPPED (no longer visible): {dropped}")
print(f"\nVISIBLE to model: {kept}")
print("\n=> The system prompt fell off the whiteboard. In production, pin")
print("   critical instructions so they are never the part that gets dropped.")
````

# Section 5 - Training (how the model learned, once)

**What is it?**
**Training** is the one-time, massively expensive process of feeding the model huge amounts of text and adjusting its internal numbers (**parameters / weights**) so it gets better at predicting the next token. After training, those weights are frozen and shipped.

**Why does it matter?**
Understanding training tells you what the model *can't* do live: it can't learn your new document just because you showed it once in a chat. Learning happens during training, not during a conversation.

**How does it work? (two stages)**
- **Pretraining** - the model reads enormous general text and learns language, facts, reasoning patterns. This is where most of the cost and time go (think many GPUs for weeks/months).
- **Fine-tuning / alignment** - a smaller follow-up step that shapes *behaviour*: making it follow instructions, be helpful, stay safe. Techniques include supervised fine-tuning and RLHF (reinforcement learning from human feedback).

**When will I use this?**
Rarely as a beginner - and that's a key insight. Most production needs are met *without* training, using prompting and RAG. You reach for fine-tuning only for specialised style/format/behaviour at scale.

### The critical distinction: training vs. inference
| | **Training** | **Inference** |
|---|---|---|
| When | Once, before release | Every time you ask a question |
| What changes | The model's weights | Nothing - weights are frozen |
| Cost profile | Huge, one-time | Small, per request |
| Analogy | Studying for an exam | Answering a question in the exam |

> **Why users don't 'train ChatGPT' every time they ask:** Each question is **inference** - the frozen model runs a forward pass and predicts tokens. Your chat does **not** update the model's weights. If it feels like it 'remembered', that's because the app re-sent earlier messages as context, not because it learned.

> **Why training is expensive:** It requires enormous datasets, thousands of specialised chips running for a long time, and large energy and engineering costs. That's why a handful of labs train frontier models and the rest of us *use* them.

---

### Instructor Explanation
*"Think of training like getting a medical degree - years of study, enormous effort, and at the end you have a doctor whose knowledge is set. Inference is that doctor seeing a patient: fast, repeatable, and it doesn't send them back to medical school each visit. When a student asks 'did I just teach the model something?' - no. You gave the doctor notes to read during the appointment. The moment the appointment ends, the notes are gone and the doctor's training is unchanged."*

### Remember This
> Training writes the model's brain once (expensive). Inference uses that frozen brain to answer (cheap). Chatting never retrains the model.

### Enterprise Perspective
A healthcare assistant almost never gets the latest drug guidance by *retraining* - that's slow and costly. Instead, new guidance is added to a **retrieval index (RAG)** so the frozen model can read it at inference time. Fine-tuning is reserved for things like enforcing a consistent clinical tone or output format across millions of responses.

### Interview Question
**Q: What's the difference between training and inference, and why does it matter for cost?**
> *Model answer:* "Training is the one-time, expensive process that sets the model's weights using large datasets and compute. Inference is running the frozen model to answer a query - cheap and repeatable. It matters because most production cost and iteration happens at inference (per-token), and you usually adapt behaviour with prompting/RAG rather than re-training."

### Common Misconceptions
- **Wrong:** "Chatting with the model trains it."
  **Right:** "Chatting is inference; weights don't change. New knowledge is added via context/RAG, not by talking to it."
- **Wrong:** "To use my own data, I must fine-tune."
  **Right:** "Usually RAG or prompting is enough; fine-tune only for behaviour/style/format at scale."

### Quick Check
1. **True/False:** Asking ChatGPT a question updates its weights. *(False - that's inference.)*
2. **MCQ:** The most expensive, one-time stage is (a) inference (b) pretraining (c) tokenization (d) prompting. *(b)*
3. **Scenario:** Your policy changed today. Fastest correct fix: retrain, or update the RAG index? *(Update the RAG index.)*


````python
# Hands-On: training vs inference classifier
# Decide whether each task changes the model's WEIGHTS (training) or just
# runs the frozen model (inference). Good interview warm-up.

tasks = [
    "A user asks the chatbot for today's weather phrasing",
    "A lab feeds 1 trillion tokens to adjust model weights",
    "You add company FAQs to a RAG index and query them",
    "A team runs RLHF to make answers more helpful",
    "Your app re-sends chat history so the bot 'remembers'",
    "You fine-tune a model on 50k support tickets",
]

def classify(task):
    t = task.lower()
    training_signals = ["adjust model weights", "rlhf", "fine-tune", "fine tune", "trillion tokens"]
    return "TRAINING" if any(sig in t for sig in training_signals) else "INFERENCE"

for task in tasks:
    print(f"{classify(task):10} <- {task}")

print("\nRule: if the model's weights change, it's TRAINING. Otherwise INFERENCE.")
````

# Section 6 - Inference & Token Prediction (how an answer is written)

**What is it?**
**Inference** is the model in action: it takes your tokens and generates a response **one token at a time**. **Token prediction** is the single step inside that loop - choosing the next token from a list of probabilities.

**Why does it matter?**
This is the mechanism behind how every answer is produced. Understand this one loop and the model stops being mysterious.

**How does it work? (the generation loop)**
1. The model reads all current tokens.
2. It outputs a **probability for every token** in its vocabulary (e.g., "Paris" 71%, "London" 6%, ...).
3. A **sampling step** picks one token from those probabilities.
4. The chosen token is appended to the text.
5. Repeat from step 1 until the model emits a stop signal.

```
Prompt: "The capital of France is"
   |
   v
[ Model ] -> probabilities:  Paris 0.71 | London 0.06 | Lyon 0.04 | ...
   |
   v
[ Sampling ] -> picks "Paris"
   |
   v
"The capital of France is Paris"  -> feed back in -> predict next token -> ...
```

**When will I use this?**
Whenever you wonder "why did I get a different answer this time?" - the answer lives in steps 2 and 3.

### Why responses differ for the same prompt
Step 3 is not always "pick the single highest-probability token." Depending on the settings, the model may *sample* among several likely tokens. Different draws -> different (but still plausible) sentences. With the most deterministic settings, you get the same answer every time.

---

### Instructor Explanation
*"Let's role-play. I'll be the model. Someone start a sentence: 'On Sunday morning I like to...' Now everyone shout a next word. Notice you didn't all say the same thing - 'sleep', 'run', 'read' - but they're all reasonable. The model has the same menu of reasonable next words, each with a probability. Picking the safe favourite every time gives predictable text; occasionally picking a tasty long-shot gives creative text. That single choice is the whole game."*

### Remember This
> Inference writes one token at a time. Each step is a weighted dice roll over likely next tokens.

### Enterprise Perspective
For a customer-support assistant, you want answers that are consistent and reproducible, so you push the sampling toward "always pick the top token" (low randomness). For a marketing copy generator, you loosen it so the same brief yields fresh variations. Same engine, different dice settings.

### Interview Question
**Q: Walk me through how an LLM generates a sentence.**
> *Model answer:* "At inference, the model reads the current tokens and outputs a probability distribution over the next token. A sampling step selects one token, which is appended to the sequence. The loop repeats, re-reading the longer sequence each step, until a stop token ends generation. Generation is autoregressive: one token at a time, each conditioned on all previous tokens."

### Common Misconceptions
- **Wrong:** "The model writes the whole answer at once."
  **Right:** "It's autoregressive - one token at a time, each based on everything so far."
- **Wrong:** "Same prompt always gives the same answer."
  **Right:** "Only with deterministic settings; sampling introduces variation by design."

### Quick Check
1. **True/False:** LLMs generate all words simultaneously. *(False - one token at a time.)*
2. **MCQ:** The randomness in 'same prompt, different answer' comes from (a) training (b) the sampling step (c) the tokenizer (d) the network cable. *(b)*
3. **Scenario:** You need byte-for-byte identical output every run. Which direction do you push randomness? *(Down - toward always picking the top token.)*


````python
# Hands-On: watch token prediction + sampling (no API needed)
import random

# Pretend the model produced these probabilities for the next token:
next_token_probs = {
    "Paris":  0.71,
    "London": 0.10,
    "Lyon":   0.08,
    "Nice":   0.06,
    "Berlin": 0.05,
}

def greedy(probs):
    # Deterministic: always pick the most likely token
    return max(probs, key=probs.get)

def sample(probs):
    # Random: draw a token according to its probability
    tokens, weights = list(probs.keys()), list(probs.values())
    return random.choices(tokens, weights=weights, k=1)[0]

print("Greedy (deterministic) pick, 5 runs:")
print("  " + ", ".join(greedy(next_token_probs) for _ in range(5)))

print("\nSampled (random) pick, 10 runs:")
print("  " + ", ".join(sample(next_token_probs) for _ in range(10)))

print("\n=> Greedy is repeatable. Sampling sometimes explores 'London'/'Lyon'.")
print("   Sampling decides how often the model explores less-likely tokens.")
````

# Section 7 - Hallucinations (why models confidently make things up)

**What is it?**
A **hallucination** is when the model produces text that is fluent and confident but **factually wrong or fabricated** - invented citations, fake figures, a made-up policy.

**Why does it matter?**
This is the single biggest risk in enterprise AI. A wrong-but-confident answer in banking, healthcare, or legal can cause real harm. Knowing *why* it happens tells you how to prevent it.

**How does it happen?**
Remember the core fact: the model predicts the *most plausible-sounding* next tokens, not the *most true* ones. When it lacks the right information, it doesn't stop - it fills the gap with something statistically likely. Common triggers:
- The fact wasn't well represented in training, or is newer than the model.
- You asked about something specific to *you* that the model was never given (your policy, your data).
- High randomness in sampling pushed it toward a less-likely, wrong token.
- The prompt invited guessing instead of allowing "I don't know."

**When will I deal with this?**
On every production system. Reducing hallucinations is a core part of an AI engineer's job.

### How production teams reduce hallucinations
- **Ground with RAG:** give the model the real source text and instruct it to answer *only* from that context.
- **Allow an escape hatch:** explicitly permit "I don't have enough information" so it isn't forced to guess.
- **Lower randomness:** use more deterministic sampling for factual tasks.
- **Constrain and verify:** structured outputs, validators, and human review for high-stakes answers.
- **Cite sources:** require the model to point at the retrieved text so claims are checkable.

> **The precise mental model:** Hallucination isn't a bug the model "decides" to commit - it's the same next-token prediction doing its job without grounding. Fluent != true. Your architecture supplies the truth.

---

### Instructor Explanation
*"Picture a very articulate student in an oral exam who didn't study one topic. Instead of saying 'I don't know', they smoothly improvise something that *sounds* right - confident tone, plausible words, completely invented. That's a hallucination. The fix isn't to make the student more confident; it's to let them say 'I don't know' and to hand them the textbook open to the right page. In AI, that open textbook is RAG."*

### Remember This
> Hallucinations happen because the model optimises for plausible-sounding, not true. Fix it by grounding it in real sources, lowering randomness, and letting it say 'I don't know.'

### Enterprise Perspective
An insurance claims assistant must never invent a clause. Production design: retrieve the actual policy text (RAG), use low-randomness settings, instruct "answer only from the provided context; if not present, say you don't know," and log sources for audit. Hallucination control is mostly *architecture and prompting*, not a single setting.

### Interview Question
**Q: Why do LLMs hallucinate, and how do you reduce it in production?**
> *Model answer:* "They hallucinate because they generate the most statistically plausible continuation, not verified truth; when grounding is missing, they fill gaps confidently. Reduce it by grounding with RAG and instructing the model to answer only from provided context, allowing an 'I don't know' response, reducing randomness, using structured outputs and validation, and citing sources for verifiability."

### Common Misconceptions
- **Wrong:** "Hallucinations mean the model is broken."
  **Right:** "It's expected behaviour of ungrounded next-token prediction; you design around it."
- **Wrong:** "Low randomness stops hallucinations."
  **Right:** "It reduces randomness but a confidently wrong answer can still be produced; grounding matters more."

### Quick Check
1. **True/False:** A confident tone means the answer is reliable. *(False.)*
2. **MCQ:** The most effective single defence against hallucination in a knowledge bot is (a) higher randomness (b) RAG grounding + 'say I don't know' (c) a longer answer (d) a longer prompt of adjectives. *(b)*
3. **Scenario:** Your bot invents a refund rule that doesn't exist. Name two fixes. *(Ground it with the real policy via RAG; allow/instruct 'I don't know'; also reduce randomness.)*


# Section 8 - Production Prompt Patterns

Now we connect everything to real systems. Each example shows a **goal**, a **prompt pattern**, and the **expected output**. These are the patterns you'll reuse constantly as an AI engineer.

> **Where the control comes from:** in these patterns the real control is **clear instructions and grounding** - telling the model exactly what to do and giving it the source text - not sampling knobs.


## 1. Factual Q&A - banking (low hallucination)
**Goal:** accurate, safe, grounded answer.
```
You are a banking assistant.
Answer ONLY from the given context.
If the answer is not in the context, say "I don't know."

Context:
Savings account interest rate is 4% per annum.

Question:
What is the savings account interest rate?
```
**Why it works:** the "answer only from context / else say I don't know" rule is the real hallucination guard - it forces the model to stay grounded.
**Expected output:**
```
The savings account interest rate is 4% per annum.
```

## 2. RAG - context-grounded answer
**Goal:** strictly use retrieved data.
```
You are an AI assistant.
Rules:
- Only use the provided context.
- Do not add external knowledge.
- Answer in 2 sentences.

Context:
The chatbot supports balance check, mini statement, and fund transfer.

Question:
What services are available?
```
**Why it works:** the grounding comes from the *rules* - "only use the provided context" keeps the answer tied to the retrieved text.
**Expected output:**
```
The chatbot supports balance check, mini statement, and fund transfer.
These are the available services mentioned in the context.
```

**Goal:** an imaginative, expressive story.
```
Write a short imaginative story about a cat who becomes a space explorer.
Make it emotional and surprising.
```
**Why it works:** an open, expressive instruction invites variety. Grounding doesn't matter here; the prompt asks for imagination.
**Expected output (sample - will vary each run, by design):**
```
The cat had always watched the stars from a broken rooftop.
One day, it didn't just watch - it left.
Drifting through the silence, it learned that loneliness feels the
```

**Goal:** correct, repeatable code.
```
Write a Java method to reverse a string.

```
**Why it works:** a tight, unambiguous instruction ("return only code, no explanation") keeps the output clean and predictable.
**Expected output:**
```java
public String reverse(String input) {
    return new StringBuilder(input).reverse().toString();
}
```

**Goal:** machine-readable output an app can parse.

```
Extract details from the text and return JSON only.

Text:
Arul is 36 years old and works at Cognizant.

Format:
{ "name": "", "age": 0, "company": "" }
```
**Why it works:** giving an explicit format/template makes the JSON predictable. In production, also validate the JSON and (if available) use the model's structured-output / JSON mode.
**Expected output:**

```
Output: SELECT * FROM orders WHERE amount > 1000;

```
**Why it works:** the examples ("few-shot") steer the format. Note: this is *in-context learning*, NOT training - the weights don't change.

**Goal:** logical clarity on a multi-step problem.
```
Solve step by step:
```

**Why it works:** asking for steps ("chain-of-thought style") improves reliability on math and logic problems.
**Expected output:**
```
The train travels 60 km in 1 hour.
In 5 hours: 60 x 5 = 300 km.
Final answer: 300 km.
```

**Goal:** safe enterprise output - refuse to invent.
```
You must not make up information.
If unsure, respond with: "I don't have enough information."

Question:

```
**Why it works:** an explicit "I don't know" escape hatch is the textbook anti-hallucination pattern - it gives the model permission not to guess.
**Expected output:**

# Day 1 - One-Page Cheat Sheet (5-minute revision)

### Key definitions
- **LLM** - neural network that predicts the next **token**; reconstructs answers from learned patterns (no live database).
- **Token** - chunk of text the model processes (often a word fragment). **~100 tokens ~ 75 English words; ~4 chars/token.**
- **Context window** - max **tokens** per request: system prompt + history + retrieved docs + reply. Overflow -> oldest tokens drop.
- **Training** - one-time, expensive; sets the weights (pretraining -> fine-tuning/RLHF).
- **Inference** - running the frozen model per request; chatting never retrains it.
- **Token prediction** - autoregressive loop: read tokens -> probabilities -> sample one -> append -> repeat.
- **Hallucination** - fluent but false output; model optimises for *plausible*, not *true*.
- **RAG** - retrieve relevant text into the context window so answers are grounded and current.

### Interview one-liners
- *Token:* unit of text the model reads; billing + context are token-based.
- *Context window:* per-request token budget shared by prompt, history, docs, and reply.
- *Training vs inference:* training sets weights once (costly); inference runs the frozen model per query.
- *Why hallucinations:* ungrounded next-token prediction favours plausible over true; fix with RAG + an "I don't know" option.
- *Inference:* generating a response from the frozen model, one token at a time.

### Production recommendations
1. Ground important answers with **RAG**; instruct "answer only from context, else say I don't know."
2. Mind the **context window**: `input + output <= window`.
3. Validate structured outputs; cite sources for high-stakes answers.

### Exam revision triggers
- "Words?" -> No, **tokens** (and they're billed).
- "Remembers me?" -> No, **stateless**; history is re-sent.
- "Learns when I chat?" -> No, that's **inference**.
- "Confident = correct?" -> No, **hallucination** risk.
